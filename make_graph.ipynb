```python
import matplotlib.pyplot as plt
import pandas as pd
import yfinance as yf  # You'll need this library to download stock data

# Download Tesla stock data
tesla_data = yf.download('TSLA', start='2020-01-01', end='2023-01-01')
tesla_data.reset_index(inplace=True)  # This converts the date index to a column

# Download GameStop stock data
gme_data = yf.download('GME', start='2020-01-01', end='2023-01-01')
gme_data.reset_index(inplace=True)  # This converts the date index to a column

# Function to make a stock vs date graph
def make_graph(data, title):
    plt.figure(figsize=(10,6))
    plt.plot(data['Date'], data['Close'], label="Close Price")
    plt.xlabel("Date")
    plt.ylabel("Stock Price (USD)")
    plt.title(title)
    plt.legend()
    plt.grid(True)
    plt.show()

# Plot Tesla stock data
make_graph(tesla_data, "Tesla Stock Price Over Time")

# Plot GameStop stock data
make_graph(gme_data, "GameStop Stock Price (2020-2023)")
```
